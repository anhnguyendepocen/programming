```python
import heapq
from operator import itemgetter

filepath = 'TextHomework3.txt'
stopwords = set([
    '\n',
    '------------------------------------------------------------\n',
    'hw0'
])


def most_common(filepath, n):
    counts = {}
    with open(filepath) as f:
        for line in f:
            words = line.split(' ')
            for word in words:
                if word not in stopwords:
                    counts[word] = counts.get(word, 0) + 1     
    
    common = heapq.nlargest(n, counts.items(), key = itemgetter(1))
    return common
```
