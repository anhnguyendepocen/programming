```python
%matplotlib inline
%load_ext watermark
%config InlineBackend.figure_format = 'retina'

import pandas as pd
import matplotlib.pyplot as plt

%watermark -a 'Ethen' -d -t -v -p pandas,matplotlib 
```

```python
# stacked bar chart requires the differences
# between the values, e.g. For Monday, B's value
# is around 11, but the value for it will be
# 11 - 3 (A's value)
A_values = 3, 10, 21, 2, 5
B_values = 8, 5, 4, 0.5, 2
C_values = 12, 16, 1, 19.5, 5
D_values = 15, 5, 5, 15, 6
E_values = 1, 12, 1.5, 15, 6
df = pd.DataFrame({
    'A': A_values,
    'B': B_values,
    'C': C_values,
    'D': D_values,
    'E': E_values})

fontsize = 12
ticklabels = 'M', 'T', 'W', 'Th', 'F'
color = '#3D50BD', '#8EB2FB', '#DDDCDC', '#F2987D', '#B20B2B'
ax = df.plot(
    kind = 'bar', stacked = True,
    fontsize = fontsize, figsize = (10, 8),
    title = 'user activity', color = color,
    edgecolor = ['black'] * len(color))
ax.grid(linestyle = 'dashed')
ax.set_yticks([0, 10, 20, 30, 40, 50, 60])
ax.set_xticklabels(ticklabels, rotation = 'horizontal')
ax.legend(fontsize = fontsize)
plt.show()
```
